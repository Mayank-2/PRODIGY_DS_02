```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy  import stats
import seaborn as sns
```

```python
Titanic_data = pd.read_csv('train.csv')
```

```python
Titanic_data.head()
```

```python
Titanic_data.info()
```

```python
Titanic_data.describe()
```

```python
Titanic_data.isnull().sum()
```

```python
Titanic_data['Age'].fillna(Titanic_data['Age'].median(),inplace=True)
```

```python
Titanic_data.drop_duplicates(inplace=True)
```

```python
Titanic_data.isnull().sum()
```

```python
Titanic_data.rename(columns={'Pclass': 'PassengerClass'}, inplace=True)
```

```python
plt.hist(Titanic_data['Age'], bins=20)
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Age Distribution')
plt.show()
```

```python
sns.scatterplot(x='Age', y='Fare', data=Titanic_data)
plt.xlabel('Age')
plt.ylabel('Fare')
plt.title('Age vs Fare')
plt.show()
```

```python
correlation_matrix = Titanic_data.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
correlation_age_fare = Titanic_data['Age'].corr(Titanic_data['Fare'])
print(f"Correlation between Age and Fare: {correlation_age_fare:.2f}")
```

# Correlation between Age and Fare: 0.10

```python
sns.barplot(x='PassengerClass', y='Fare', data=Titanic_data)
plt.xlabel('Passenger Class')
plt.ylabel('Fare')
plt.title('Fare by Passenger Class')
plt.show()
```
